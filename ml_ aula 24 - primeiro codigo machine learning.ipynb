```python
import pandas as pd
arquivo = pd.read_csv("D:/Pública/wine_dataset.csv")
```

```python
arquivo.head()
```

```python
# quais as caracteristicas que tornam o vinho ser tinto ou branco
```

```python
# Transformar o "style" de formato de texto pra formato numérico
arquivo["style"] = arquivo["style"].replace("red",0) # substitui o que for red por ZERO
```

```python
arquivo["style"] = arquivo["style"].replace("white",1) # substitui o que for white por UM
```

```python
arquivo # mostrar que foi substituido
```

```python
# Separando as variáveis entre preditoras e variável alvo
y = arquivo["style"] # variável alvo
x = arquivo.drop("style", axis = 1) # Excluindo a coluna "style". axis = 1 significa que é coluna. Todo o arquivo tá sendo armazenado em x, exceto a 
# coluna "style" que está sendo excluida.
```

```python
y # só o "style"
```

```python
x # Todo o arquivo, exceto o "style"
```

```python
from sklearn.model_selection import train_test_split

# criando os conjuntos de dados de treino e teste:
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) # 0.3 = Escolhe aleatoriamente 30% dos dados pra teste e 30% dos dados pra treino
```

```python
y_treino.shape # número de amostras
```

```python
x_treino.shape # número de amostras
```

```python
y_teste.shape # número de amostras
```

```python
x_teste.shape # número de amostras
```

```python
from sklearn.ensemble import ExtraTreesClassifier
# Criação do modelo: Vai pegar os dados de treino pra aprender
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino) # fit -> Vou aplicar o algoritmo de machine learning nos meus dados de treino

# Imprimindo resultados
resultado = modelo.score(x_teste, y_teste) # score vai pegar conjunto de dados de teste e vai comparar com o y_teste (que são os resultados reais) e 
# vai tentar prever qual é a classe (Se é a classe ZERO ou UM)
print("Acurácia: ", resultado)
```
